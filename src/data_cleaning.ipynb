```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/python_faq_dataset.csv')
print(df.head())
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
```

```python
print("=== BEFORE CLEANING ===")
print("Total rows:", len(df))
print("\nMissing values:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Fill missing values
df['score'] = df['score'].fillna(0)
df['tags'] = df['tags'].fillna('unknown')

print("=== AFTER CLEANING ===")
print(df.isnull().sum())
```

```python
df.to_csv('../data/cleaned_python_faq.csv', index=False)
print("Saved successfully!")
```
